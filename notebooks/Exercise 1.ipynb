# Exercise 1: bkmeans

In this assignment, agglomerative clustering will be implemented and demonstrated on two toy data sets.

```python
import matplotlib.pyplot as plt
from models.BKMeans import bkmeans
from sklearn.datasets import make_multilabel_classification as make_ml_clf
from sklearn.datasets import make_circles
```

```python
ml_X, _ = make_ml_clf(n_samples=200, n_features=2, n_classes=4, n_labels=1)
circle_X, _ = make_circles(n_samples=200, noise=0.1, factor=0.1)
```

```python
ml_clustering = bkmeans(ml_X, 3, 5)
circle_clustering = bkmeans(circle_X, 3, 5)
```

```python
fig, axes = plt.subplots(2)
axes[0].scatter(ml_X[:, 0], ml_X[:, 1], c=ml_clustering)
axes[1].scatter(circle_X[:, 0], circle_X[:, 1], c=circle_clustering)
plt.show()
```
